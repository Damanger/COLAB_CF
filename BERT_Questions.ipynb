```python
import requests
from bs4 import BeautifulSoup  ## Procesar codigo HTML
from transformers import pipeline
```

```python
# --- ETAPA 1: Obtener el contenido de la página web ---
def extract_text_from_url(url):
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception(f"Error al obtener la página. Código: {response.status_code}")

    soup = BeautifulSoup(response.content, 'html.parser')
    paragraphs = soup.find_all('p')  # Extraer todos los párrafos <p>

    # Concatenar el texto de los párrafos
    text = ' '.join([para.get_text() for para in paragraphs])
    return text

# URL de Wikipedia
url = "https://en.wikipedia.org/wiki/List_of_cloud_types"
url = "https://en.wikipedia.org/wiki/Oaxaca"
url = "https://www.bbc.com/news/articles/cn8g9pzn8z0o"
```

```python
try:
    print("Extrayendo texto de la página web...")
    document_text = extract_text_from_url(url)
    print("Texto extraído correctamente.")
except Exception as e:
    print(e)
    exit()
```

```python
len(document_text), len(document_text.split(" "))
document_text = document_text[:2000]
```

```python
document_text = "Russian missile attack kills eight in Zaporizhzhia A Russian missile attack has killed at least eight people and injured another 22 - including a child - in Ukraine's southern city of Zaporizhzhia, local officials say. As many as five people may still be trapped under rubble after Tuesday's strike on a private clinic and residential buildings in the city centre, police say. Overnight, rescuers pulled out two women from the wreckage. They are now being treated in hospital. A search and rescue operation is continuing. Zaporizhzhia regional head Ivan Fedorov says Russia fired a ballistic missile, most likely an Iskander. Russia's defence ministry has not commented.  "
```

```python
#  Configurar el modelo de BERT para QA ---

qa_pipeline = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")
```

```python
question = "how many people were injured?"
# Ejecutar el modelo para responder la pregunta
answer = qa_pipeline({
  'question': question,
  'context': document_text
})
print(f"Respuesta: {answer['answer']} (Confianza: {answer['score']:.2f})\n")
```
